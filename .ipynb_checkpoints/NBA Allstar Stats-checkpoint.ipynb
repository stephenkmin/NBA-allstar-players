```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Pulling in the all star game list from csv file
allstar_file = "Resources/NBA_All_Star_Games.csv"
allstar_df = pd.read_csv(allstar_file)
allstar_df.head()
```

```python
# Changing the year data to fit seasons
allstar_df.loc[allstar_df.Year == 2016, 'Year'] = "2015 - 2016"
allstar_df.loc[allstar_df.Year == 2015, 'Year'] = "2014 - 2015"
allstar_df.loc[allstar_df.Year == 2014, 'Year'] = "2013 - 2014"
allstar_df.loc[allstar_df.Year == 2013, 'Year'] = "2012 - 2013"
allstar_df.loc[allstar_df.Year == 2012, 'Year'] = "2011 - 2012"
allstar_df.loc[allstar_df.Year == 2011, 'Year'] = "2010 - 2011"
allstar_df.loc[allstar_df.Year == 2010, 'Year'] = "2009 - 2010"
allstar_df.loc[allstar_df.Year == 2009, 'Year'] = "2008 - 2009"
allstar_df.loc[allstar_df.Year == 2008, 'Year'] = "2007 - 2008"
allstar_df.loc[allstar_df.Year == 2007, 'Year'] = "2006 - 2007"
allstar_df.loc[allstar_df.Year == 2006, 'Year'] = "2005 - 2006"
allstar_df.loc[allstar_df.Year == 2005, 'Year'] = "2004 - 2005"
allstar_df.loc[allstar_df.Year == 2004, 'Year'] = "2003 - 2004"
allstar_df.loc[allstar_df.Year == 2003, 'Year'] = "2002 - 2003"
allstar_df.loc[allstar_df.Year == 2002, 'Year'] = "2001 - 2002"
allstar_df.loc[allstar_df.Year == 2001, 'Year'] = "2000 - 2001"
allstar_df.loc[allstar_df.Year == 2000, 'Year'] = "1999 - 2000"
allstar_df.head()
```

```python
# Adding a column for to id each row
allstar_df['id'] = allstar_df.index
allstar_df.head()
```

```python
# Renaming the columns and setting the index
allstar_df = allstar_df[["id", "Year", "Player"]]
allstar_df = allstar_df.rename(columns={"Year" : "season",
                                        "Player" : "player"})

allstar_df.set_index("id", inplace=True)
allstar_df.head()
```

```python
# Pulling in the player stats by season from csv file
stats_file = "Resources/players_stats_by_season_full_details.csv"
stats_df = pd.read_csv(stats_file)
stats_df.head()
```

```python
# Adding a column for to id each row
stats_df['id'] = stats_df.index
stats_df.head()
```

```python
# Dropping playoff stats
drop_playoff = stats_df[ stats_df["Stage"] == "Playoffs" ].index
stats_df.drop(drop_playoff , inplace=True)

# Renaming the columns and setting the index
stats_df = stats_df[["id", "Season", "GP", "Player", "PTS", "REB", "AST", "STL", "BLK", "TOV"]]
stats_df = stats_df.rename(columns={"Season" : "season",
                                    "GP" : "games_played",
                                    "Player" : "player",
                                    "PTS" : "points",
                                    "REB" : "rebounds",
                                    "AST" : "assists",
                                    "STL" : "steals",
                                    "BLK" : "blocks",
                                    "TOV" : "turnovers"})
stats_df.set_index("id", inplace=True)
stats_df.head()
```

```python
# Changed all the stats to reflect per games played
stats_df["points"] = round(stats_df["points"]/stats_df["games_played"], 2)
stats_df["rebounds"] = round(stats_df["rebounds"]/stats_df["games_played"], 2)
stats_df["assists"] = round(stats_df["assists"]/stats_df["games_played"], 2)
stats_df["steals"] = round(stats_df["steals"]/stats_df["games_played"], 2)
stats_df["blocks"] = round(stats_df["blocks"]/stats_df["games_played"], 2)
stats_df["turnovers"] = round(stats_df["turnovers"]/stats_df["games_played"], 2)
stats_df.head()
```

```python
allstar_df.to_sql(name='allstar', con=engine, if_exists='append', index=True)
```

```python
stats_df.to_sql(name='player_stats', con=engine, if_exists='append', index=True)
```
